```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split #splitting the data into train and test
from sklearn.preprocessing import StandardScaler # standard preprocesing
from sklearn.neighbors import KNeighborsClassifier # Algorithm for knn
from sklearn.metrics import confusion_matrix #testing the model
from sklearn.metrics import f1_score #testing the model
from sklearn.metrics import accuracy_score #testing the model

```

```python
dataset = pd.read_csv(r'C:\Users\alber\OneDrive - INTEC\LAB MACHINE LEARNING\P2\diabetes.csv')
print(len(dataset))
print(dataset.head())
```

```python
# Replace the missing or not acceptable values with NaN and after with the mean
zero_not_accepted = ['Glucose','BloodPressure','SkinThickness','BMI','Insulin']
for column in zero_not_accepted:
    dataset[column]=dataset[column].replace(0,np.NaN)
    mean = int(dataset[column].mean(skipna=True))
    dataset[column]= dataset[column].replace(np.NaN, mean)
    
```

```python
print(dataset['Glucose'])
```

```python
#Split dataset into train and test
X = dataset.iloc[:,0:8]
y = dataset.iloc[:,8]
X_train,X_test, y_train, y_test = train_test_split(X,y,random_state=0, test_size=0.2)#test size significa que cogeremos un 20% de la muestra en este caso y lo probaremos despues
print("Trained X \n",X_train,"\n","Test X \n",X_test,"\n","Trained y \n",y_train,"\n","test y \n",y_test)
```

```python
#Feature Scaling, lo estandariza para que toda la data este en el mismo rango en este caso de -1 a 1
sc_X= StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)

print("Trained X \n",X_train)
print("Test X \n", X_test)
```

```python
#Lenght'
import math as m
m.sqrt(len(y_test))
```

```python
#Define the model: INIT K-NN
classifier = KNeighborsClassifier(n_neighbors=11,p=2,metric='euclidean')
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
#Evaluate the model
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
print(f1_score(y_test,y_pred))
```

```python
#accuracy of the model
print(accuracy_score(y_test,y_pred)*100,"%")
```

```python
import numpy as np

def euclidean_distance(x1, x2):
    """Compute the Euclidean distance between two points."""
    return np.sqrt(np.sum((x1 - x2)**2))

def condensed_nearest_neighbors(X, y):
    """Condensed Nearest Neighbors (CNN) algorithm."""
    condensed_X = [X[0]]  # Initialize condensed dataset with the first instance
    condensed_y = [y[0]]  # Corresponding labels

    for i in range(1, len(X)):
        x = X[i]
        label = y[i]

        # Check if the instance can be correctly classified by the condensed dataset
        classified_correctly = False
        for j in range(len(condensed_X)):
            if condensed_y[j] == label and euclidean_distance(x, condensed_X[j]) == 0:
                classified_correctly = True
                break
        
        # If not, add it to the condensed dataset
        if not classified_correctly:
            condensed_X.append(x)
            condensed_y.append(label)

    return np.array(condensed_X), np.array(condensed_y)

# Example usage:
# Generate some random data as a training set
np.random.seed(0)
X = np.random.rand(10000000, 2)
y = np.random.randint(0, 2, 10000000)

# Apply CNN
condensed_X, condensed_y = condensed_nearest_neighbors(X, y)

print("Original dataset size:", X.shape)
print("Condensed dataset size:", condensed_X.shape)
```
